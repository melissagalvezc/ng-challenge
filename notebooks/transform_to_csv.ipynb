```python
import os
import pandas as pd

# PATH
txt_folder = './files'
csv_folder = './files/csv'


# ITERATE THROUGH THE FILES IN THE TEXT FOLDER
for filename in os.listdir(txt_folder):
    if filename.endswith('.txt') and 'ce.' in filename:
        txt_path = os.path.join(txt_folder, filename)
        base_name = filename.replace('.txt', '')
        csv_filename = base_name.replace('.', '_')
        csv_path = os.path.join(csv_folder, csv_filename)+'.csv'
        try:
            df = pd.read_csv(txt_path, sep="\t")
            df.to_csv(csv_path, index=False)
            print(f"File {csv_filename} created.")
        except Exception:
            print(f"Error:  file {csv_filename}")
            
```
